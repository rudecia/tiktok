```python
import numpy as np

#these should all be on a scale from 0 to 10
#s: sexual ability
#p: personality/emotional intelligence
#v: value alignment
#b: baseline wellbeing

#i, j, k should correspond to relative importance of s, p, and v respectively
#i, j, and k should add to 1

def relationship_worth(s, p, v, b, i, j, k):
  ranking_vector = np.array([s, p, v])
  importance_vector = np.array([i,j,k])
  score = int(b > 4) * np.dot(ranking_vector, importance_vector) - b

  if score < 0:
    return('not worth it')
  elif score > 0 and score < 2:
    return('worth it')
  else:
    return('marriage material')




```

```python
relationship_worth(7,5,2,7, 0.2,0.4,0.4)
```
